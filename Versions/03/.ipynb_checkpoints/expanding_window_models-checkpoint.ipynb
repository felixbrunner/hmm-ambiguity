## Model

### Theory
The rationale for the model goes as follows:

Let $r^i_{t}$ and $r^M_{t}$ denote excess returns on an asset $i$ and the market portfolio. 

Investors use a standard CAPM to form expectations about means and variances of asset excess returns:
$$\mathbb{E}_t\big[r^i_{t+1}\big] = \alpha^i + \beta^i \, \mathbb{E}_t\big[r^{M}_{t+1}\big]$$
$$\mathbb{V}_t\big[r^i_{t+1}\big] = (\beta^i)^2\times(\sigma^M)^2+(\sigma^i)^2$$

Now suppose agents are ambiguous about the set of model-parameters that are generating next-period returns. In particular, there are two competing sets of parameters $\theta^i_{S_{t+1}} = \big(\alpha^i_{S_{t+1}}, \beta^i_{S_{t+1}}, \sigma^i_{S_{t+1}}\big)$, where $S_{t+1} \in \{0,1\}$, and the transition between the two sets of parameters follows a Markov process:

\begin{equation*}
    \begin{cases}
        r^i_{t+1} = \alpha^i_{S_{t+1}=0} + \beta^i_{S_{t+1}=0} \, \mathbb{E}_t\big[r^M_{t+1}\big] + \varepsilon^i_{t+1},  \quad\quad   \varepsilon^i_{t+1} \sim \mathcal{N}\big(0,(\sigma^i_{S_{t+1}=0})^2\big)\\
        r^i_{t+1} = \alpha^i_{S_{t+1}=1} + \beta^i_{S_{t+1}=1} \, \mathbb{E}_t\big[r^M_{t+1}\big] + \varepsilon^i_{t+1},  \quad\quad   \varepsilon^i_{t+1} \sim \mathcal{N}\big(0,(\sigma^i_{S_{t+1}=1})^2\big)
    \end{cases}
\end{equation*}

\begin{equation*}
    P\big(S^i_{t+1} = s^i_{t+1} \, \big| \, S^i_{t} = s^i_{t}\big) = 
    \begin{bmatrix}
        p^i_{00} & 1-p^i_{00}\\
        p^i_{10} & 1-p^i_{10}\\
    \end{bmatrix}
\end{equation*}

Given the two-regime markov process, investors are ambiguous between two Gaussian distributions that next-period returns could be drawn from:
\begin{equation}
    \begin{cases}
        \phi_0(r^i_{t+1}) = \mathcal{N}\big(\mu^i_0 = \alpha^i_{0} + \beta^i_{0} \, \mathbb{E}_t\big[r^M_{t+1}\big],(\sigma^i_{0})^2\big), \quad \text{with probability } P_t(S^i_{t+1} = 0)\\
        \phi_1(r^i_{t+1}) = \mathcal{N}\big(\mu^i_1 = \alpha^i_{1} + \beta^i_{1} \, \mathbb{E}_t\big[r^M_{t+1}\big],(\sigma^i_{1})^2\big), \quad \text{with probability } P_t(S^i_{t+1} = 1) = \big[1-P_t(S^i_{t+1} = 0)\big]
    \end{cases}
\end{equation}
For both distributions, $\beta^i_{S_{t+1}}$ reflects the state-dependent level of systematic risk, while $(\sigma^i_{S_{t+1}})^2$ denotes the asset's state-dependent level of idiosyncratic risk.

With knowledge of regime-probabilities at time $t$, next-period $t+1$ regime-probabilities are given by:
\begin{equation*}
    \begin{bmatrix}
        P_t(S^i_{t+1} = 0)&
        P_t(S^i_{t+1} = 1)\\
    \end{bmatrix}=
    \begin{bmatrix}
        P_t(S^i_{t} = 0)&
        P_t(S^i_{t} = 1)\\
    \end{bmatrix}
    \begin{bmatrix}
        p^i_{00} & 1-p^i_{00}\\
        p^i_{10} & 1-p^i_{10}\\
    \end{bmatrix}
\end{equation*}

Note that although both priors rely on normality assumptions, the weighted expectiation about the next-period return distribution (by probability-weighting the two priors) is no longer normal. Instead, it follows a mixture distribution $\mathcal{D}\big(\mu^i_D,(\sigma^i_D)^2,.\big)$ with first and second moments $\mu^i_D$ and $(\sigma^i_D)^2$, and finite higher moments:
\begin{align}
    \mathbb{E}_t\big[\phi(r^i_{t+1})\big]
    &= P_t(S^i_{t+1} = 0) \times \phi_0(r^i_{t+1}) + \big[1-P_t(S^i_{t+1} = 0)\big] \times \phi_1(r^i_{t+1})\\
    &= P_t(S^i_{t+1} = 0) \times \mathcal{N}\big(\mu^i_0,(\sigma^i_{0})^2\big) + \big[1-P(S^i_{t+1} = 0)\big] \times \mathcal{N}\big(\mu^i_1,(\sigma^i_{1})^2\big)\\
    %&= N\big(P(S^i_{t+1} = 0) \times \big(\alpha^i_{0} + \beta^i_{0} \times \mathbb{E}_t\big[r^M_{t+1}\big]\big) + \big[1-P_t(S^i_{t+1} = 0)\big] \times \big(\alpha^i_{1} + \beta^i_{1} \times \mathbb{E}_t\big[r^M_{t+1}\big]\big), \quad P(S^i_{t+1} = 0) \times (\sigma^i_{0})^2 + \big[1-P_t(S^i_{t+1} = 0)\big] \times (\sigma^i_{0})^2 \big)
    %&= \mathcal{D}\big(P(S^i_{t+1} = 0) \times \big(\alpha^i_{0} + \beta^i_{0} \times \mathbb{E}_t\big[r^M_{t+1}\big]\big)+ \big[1-P(S^i_{t+1} = 0)\big] \times \big(\alpha^i_{1} + \beta^i_{1} \times \mathbb{E}_t\big[r^M_{t+1}\big]\big)\big)\\
    %&=  \mathcal{D}\big(\alpha^i_{w} + \beta^i_{w} \times \mathbb{E}_t\big[r^M_{t+1}\big]\big)
    &= \mathcal{D}\big(\mu^i_D,(\sigma^i_D)^2,.\big)
\end{align}

The first two moments of $\mathcal{D}(.)$ are given by:
\begin{align*}
\mu^i_D &= \sum_{s_{t+1}=0}^{1} P_t(S^i_{t+1} = s_{t+1}) \times \mu^i_{s_{t+1}} &= \mathbb{E}_t\big[r^i_{t+1}\big] \\
(\sigma^i_D)^2 &= \sum_{s_{t+1}=0}^{1} P_t(S^i_{t+1} = s_{t+1}) \times (\sigma^i_{s_{t+1}})^2 + \sum_{s_{t+1}=0}^{1} P_t(S^i_{t+1} = s_{t+1}) \times (\mu^i_{s_{t+1}})^2 - \Big(\sum_{s_{t+1}=0}^{1} P_t(S^i_{t+1} = s_{t+1}) \times \mu^i_{s_{t+1}}\Big)^2 &= \mathbb{V}_t\big[r^i_{t+1}\big]
\end{align*}

Note that these moments can be interpreted as the expected return and expected total variance of asset $i$'s return.

### Estimation

Empirically, we can estimate a Markov regime switching model in historical return data for each asset $i$. Intercept $\alpha^i_{S_t}$, market beta $\beta^i_{S_t}$ and error variance $(\sigma^i_{S_t})^2$ are state-dependent, while transition probabilities $p_{00}$ and $p_{10}$ are static. More formally:

\begin{equation*}
   r^i_t = \alpha^i_{S_t} + \beta^i_{S_t} \, r_t^M  + \varepsilon^i_{t},  \quad\quad   \varepsilon^i_{t} \sim \mathcal{N}\big(0,(\sigma^i_{S_t})^2\big),  \quad\quad   S_t \in \{0,1\}
\end{equation*}

\begin{equation*}
    P_{t-1}\big(S^i_{t} = s^i_{t} \, \big| \, S^i_{t-1} = s^i_{t-1}\big) = 
    \begin{bmatrix}
        p^i_{00} & 1-p^i_{00}\\
        p^i_{10} & 1-p^i_{10}\\
    \end{bmatrix}
\end{equation*}

The parameters $p^i_{00}, p^i_{10}, \alpha^i_0, \alpha^i_1, \beta^i_0, \beta^i_1, (\sigma^i_{1})^2, (\sigma^i_{2})^2$ and their standard errors are estimated by maximum likelihood. 

Restrictions of the general model such as $\alpha^i_0 = \alpha^i_1$ or $(\sigma^i_{0})^2 = (\sigma^i_{1})^2$ are possible, but the current version does not restrict the general model.

From the estimation, we also obtain filtered probabilities and smoothing probabilites to be in either regime for all in-sample periods.

In addition, a standard linear regression model (standard CAPM) for each asset and data window is also estimated:
$$r^i_t = \alpha^i+\beta^i \, r^M_t, \quad\quad \varepsilon^i_t \sim \mathcal{N}\big(0,(\sigma^i)^2\big)$$
This can serve as a benchmark and help distinguish periods in which a two-state specification really does offer a better description of the data.

## Full sample models

To first analyse if this approach yields any meaningful results, we fist estimate the HMM CAPM specification in the full sample starting from January 1960.

```python
data_full_sample = pd.DataFrame(index=industries_monthly.columns)
```

```python
# This dictionary contains the mapping of data variables to model outputs for the OLS CAPM models
OLS = {'likelihood': 'model.llf',
                'aic': 'model.aic',
                'bic': 'model.bic',
                'alpha': 'model.params[\'const\']',
                'beta': 'model.params[\'Mkt-RF\']',
                'sigma2': 'model.mse_resid',
                'alpha_se': 'model.bse[\'const\']',
                'beta_se': 'model.bse[\'Mkt-RF\']',
                'nobs': 'model.nobs',
            }

# Add data columns to dataframe
data_full_sample = data_full_sample.join(pd.DataFrame(np.nan, columns=pd.MultiIndex.from_product([['OLS'], OLS.keys()]), index=data_full_sample.index))
```

```python
# This dictionary contains the mapping of data variables to model outputs for the unrestricted HMM models
HMM_full = {'convergence': 'model.mle_retvals[\'converged\']',
                'likelihood': 'model.llf',
                'aic': 'model.aic',
                'bic': 'model.bic',
                'alpha_0': 'model.params[\'const[0]\']',
                'alpha_1': 'model.params[\'const[1]\']',
                'beta_0': 'model.params[\'x1[0]\']',
                'beta_1': 'model.params[\'x1[1]\']',
                'sigma2_0': 'model.params[\'sigma2[0]\']',
                'sigma2_1': 'model.params[\'sigma2[1]\']',
                'p00': 'model.params[\'p[0->0]\']',
                'p10': 'model.params[\'p[1->0]\']',
                'alpha_0_se': 'model.bse[\'const[0]\']',
                'alpha_1_se': 'model.bse[\'const[1]\']',
                'beta_0_se': 'model.bse[\'x1[0]\']',
                'beta_1_se': 'model.bse[\'x1[1]\']',
                'sigma2_0_se': 'model.bse[\'sigma2[0]\']',
                'sigma2_1_se': 'model.bse[\'sigma2[1]\']',
                'p00_se': 'model.bse[\'p[0->0]\']',
                'p10_se': 'model.bse[\'p[1->0]\']',
                'alpha_cov': 'model.cov_params()[\'const[0]\'][\'const[1]\']',
                'beta_cov': 'model.cov_params()[\'x1[0]\'][\'x1[1]\']',
                'sigma2_cov': 'model.cov_params()[\'sigma2[0]\'][\'sigma2[1]\']',
                'prob_cov': 'model.cov_params()[\'p[0->0]\'][\'p[1->0]\']',
                'smooth_prob_0': 'model.smoothed_marginal_probabilities.iloc[-1,0]',
                'smooth_prob_1': 'model.smoothed_marginal_probabilities.iloc[-1,1]',
                'filt_prob_0': 'model.filtered_marginal_probabilities.iloc[-1,0]',
                'filt_prob_1': 'model.filtered_marginal_probabilities.iloc[-1,1]',
                'mean_prob_0': 'model.smoothed_marginal_probabilities.iloc[:,0].mean()',
                'mean_prob_1': 'model.smoothed_marginal_probabilities.iloc[:,1].mean()',
                'smooth_prob_innov': 'abs(model.smoothed_marginal_probabilities.iloc[-1,0]-model.smoothed_marginal_probabilities.iloc[-2,0])',
                'filt_prob_innov': 'abs(model.filtered_marginal_probabilities.iloc[-1,0]-model.filtered_marginal_probabilities.iloc[-2,0])',
                'nobs': 'model.nobs',
            }

# Add data columns to dataframe
data_full_sample = data_full_sample.join(pd.DataFrame(np.nan, columns=pd.MultiIndex.from_product([['HMM'], HMM_full.keys()]), index=data_full_sample.index))
```

```python
%%time
# load independent variable
x = factors_monthly['Mkt-RF'].loc[industries_monthly.index]

# asset selection
for iAsset in industries_monthly.columns:
        
    # extract industry data
    y = industries_monthly.loc[:, iAsset]
    
    # estimate OLS CAPM model
    model = sm.OLS(y, sm.add_constant(x), missing='drop').fit()
            
    for iEntry in OLS.keys():
        data_full_sample.loc[iAsset,('OLS', iEntry)] = eval(OLS[iEntry])
            
    # save estimates as starting values for HMM models
    a = model.params['const']
    b = model.params['Mkt-RF']
    s = model.mse_resid
            
    del model
        
    # HMM model 1 (full model)
    init = [0.5, 0.5, a, a, 0.99*b, 1.01*b, s, s]
    model = sm.tsa.MarkovRegression(y, k_regimes=2, exog=x, switching_trend=True, switching_exog=True, switching_variance=True, missing='drop')\
        .fit(start_params=init, cov_type='robust', method='powell')
            
    for iEntry in HMM_full.keys():
        data_full_sample.loc[iAsset,('HMM', iEntry)] = eval(HMM_full[iEntry])
                
    del model, init
```

```python
crit = 1.96

# flag non-ergodic markov chains and periodic markov chains (return to a given state after 2 periods with certainty)
data_full_sample['HMM','ergodicity'] = (data_full_sample['HMM','p00'] <= 0.9999) & (data_full_sample['HMM','p10'] >= 0.0001)
data_full_sample['HMM','aperiodicity'] = (data_full_sample['HMM','p00'] >= 0.0001) & (data_full_sample['HMM','p10'] <= 0.9999)
    
# expected duration
data_full_sample['HMM','exp_duration_0'] = 1/(1-data_full_sample['HMM','p00'])
data_full_sample['HMM','exp_duration_1'] = 1/(data_full_sample['HMM','p10'])
data_full_sample['HMM','exp_regime_duration'] = (data_full_sample['HMM','exp_duration_0']+data_full_sample['HMM','exp_duration_1'])/2

# fit comparison statistics
data_full_sample['stats','likelihood_better'] = data_full_sample[('OLS','likelihood')]<data_full_sample[('HMM','likelihood')]
data_full_sample['stats','aic_better'] = data_full_sample[('OLS','aic')]>data_full_sample[('HMM','aic')]
data_full_sample['stats','bic_better'] = data_full_sample[('OLS','bic')]>data_full_sample[('HMM','bic')]

# OLS parameter t-stats
data_full_sample['OLS','alpha_t'] = data_full_sample['OLS','alpha']/data_full_sample['OLS','alpha_se']
data_full_sample['OLS','beta_t'] = data_full_sample['OLS','beta']/data_full_sample['OLS','beta_se']

# OLS parameter significance
data_full_sample['OLS','alpha_significant'] = (data_full_sample['OLS','alpha_t']).abs() > crit
data_full_sample['OLS','beta_significant'] = (data_full_sample['OLS','beta_t']).abs() > crit

# HMM parameter t-stats
data_full_sample['HMM','alpha_0_t'] = data_full_sample['HMM','alpha_0']/data_full_sample['HMM','alpha_0_se']
data_full_sample['HMM','alpha_1_t'] = data_full_sample['HMM','alpha_1']/data_full_sample['HMM','alpha_1_se']
data_full_sample['HMM','beta_0_t'] = data_full_sample['HMM','beta_0']/data_full_sample['HMM','beta_0_se']
data_full_sample['HMM','beta_1_t'] = data_full_sample['HMM','beta_1']/data_full_sample['HMM','beta_1_se']
data_full_sample['HMM','sigma2_0_t'] = data_full_sample['HMM','sigma2_0']/data_full_sample['HMM','sigma2_0_se']
data_full_sample['HMM','sigma2_1_t'] = data_full_sample['HMM','sigma2_1']/data_full_sample['HMM','sigma2_1_se']
data_full_sample['HMM','p00_t'] = (data_full_sample['HMM','p00']-0.5)/data_full_sample['HMM','p00_se']
data_full_sample['HMM','p10_t'] = (data_full_sample['HMM','p00']-0.5)/data_full_sample['HMM','p10_se']

# HMM parameter significance
data_full_sample['HMM','alpha_0_significant'] = (data_full_sample['HMM','alpha_0_t']).abs() > crit
data_full_sample['HMM','alpha_1_significant'] = (data_full_sample['HMM','alpha_1_t']).abs() > crit
data_full_sample['HMM','beta_0_significant'] = (data_full_sample['HMM','beta_0_t']).abs() > crit
data_full_sample['HMM','beta_1_significant'] = (data_full_sample['HMM','beta_1_t']).abs() > crit
data_full_sample['HMM','sigma2_0_significant'] = (data_full_sample['HMM','sigma2_0_t']).abs() > crit
data_full_sample['HMM','sigma2_1_significant'] = (data_full_sample['HMM','sigma2_1_t']).abs() > crit
data_full_sample['HMM','p00_significant'] = (data_full_sample['HMM','p00_t']).abs() > crit
data_full_sample['HMM','p10_significant'] = (data_full_sample['HMM','p10_t']).abs() > crit

# HMM parameter t-stats against OLS alternative
data_full_sample['HMM','alpha_0_vs_ols_t'] = (data_full_sample['HMM','alpha_0']-data_full_sample['OLS','alpha'])/data_full_sample['HMM','alpha_0_se']
data_full_sample['HMM','alpha_1_vs_ols_t'] = (data_full_sample['HMM','alpha_1']-data_full_sample['OLS','alpha'])/data_full_sample['HMM','alpha_1_se']
data_full_sample['HMM','beta_0_vs_ols_t'] = (data_full_sample['HMM','beta_0']-data_full_sample['OLS','beta'])/data_full_sample['HMM','beta_0_se']
data_full_sample['HMM','beta_1_vs_ols_t'] = (data_full_sample['HMM','beta_1']-data_full_sample['OLS','beta'])/data_full_sample['HMM','beta_1_se']
data_full_sample['HMM','sigma2_0_vs_ols_t'] = (data_full_sample['HMM','sigma2_0']-data_full_sample['OLS','sigma2'])/data_full_sample['HMM','sigma2_0_se']
data_full_sample['HMM','sigma2_1_vs_ols_t'] = (data_full_sample['HMM','sigma2_1']-data_full_sample['OLS','sigma2'])/data_full_sample['HMM','sigma2_1_se']

# HMM parameter significance against OLS alternative
data_full_sample['HMM','alpha_0_vs_ols_significant'] = (data_full_sample['HMM','alpha_0_vs_ols_t']).abs() > crit
data_full_sample['HMM','alpha_1_vs_ols_significant'] = (data_full_sample['HMM','alpha_1_vs_ols_t']).abs() > crit
data_full_sample['HMM','beta_0_vs_ols_significant'] = (data_full_sample['HMM','beta_0_vs_ols_t']).abs() > crit
data_full_sample['HMM','beta_1_vs_ols_significant'] = (data_full_sample['HMM','beta_1_vs_ols_t']).abs() > crit
data_full_sample['HMM','sigma2_0_vs_ols_significant'] = (data_full_sample['HMM','sigma2_0_vs_ols_t']).abs() > crit
data_full_sample['HMM','sigma2_1_vs_ols_significant'] = (data_full_sample['HMM','sigma2_1_vs_ols_t']).abs() > crit

# HMM parameter spreads
data_full_sample['HMM','alpha_spread'] = (data_full_sample['HMM','alpha_0']-data_full_sample['HMM','alpha_1']).abs()
data_full_sample['HMM','beta_spread'] = (data_full_sample['HMM','beta_0']-data_full_sample['HMM','beta_1']).abs()
data_full_sample['HMM','sigma2_spread'] = (data_full_sample['HMM','sigma2_0']-data_full_sample['HMM','sigma2_1']).abs()
data_full_sample['HMM','prob_spread'] = (data_full_sample['HMM','p00']+data_full_sample['HMM','p10']-1).abs()

# HMM parameter spread t-stats
data_full_sample['HMM','alpha_spread_t'] = data_full_sample['HMM','alpha_spread']/ (data_full_sample['HMM','alpha_0_se']**2 + data_full_sample['HMM','alpha_1_se']**2 - 2*data_full_sample['HMM','alpha_cov'])**0.5
data_full_sample['HMM','beta_spread_t'] = data_full_sample['HMM','beta_spread']/ (data_full_sample['HMM','beta_0_se']**2 + data_full_sample['HMM','beta_1_se']**2 - 2*data_full_sample['HMM','beta_cov'])**0.5
data_full_sample['HMM','sigma2_spread_t'] = data_full_sample['HMM','sigma2_spread']/ (data_full_sample['HMM','sigma2_0_se']**2 + data_full_sample['HMM','sigma2_1_se']**2 - 2*data_full_sample['HMM','sigma2_cov'])**0.5
data_full_sample['HMM','prob_spread_t'] = data_full_sample['HMM','prob_spread']/ (data_full_sample['HMM','p00_se']**2 + data_full_sample['HMM','p10_se']**2 - 2*data_full_sample['HMM','prob_cov'])**0.5

# HMM parameter spread significance
data_full_sample['HMM','alpha_spread_significant'] = (data_full_sample['HMM','alpha_spread_t']).abs() > crit
data_full_sample['HMM','beta_spread_significant'] = (data_full_sample['HMM','beta_spread_t']).abs() > crit
data_full_sample['HMM','sigma2_spread_significant'] = (data_full_sample['HMM','sigma2_spread_t']).abs() > crit
data_full_sample['HMM','prob_spread_significant'] = (data_full_sample['HMM','prob_spread_t']).abs() > crit
```

### Summary
The estimation of the 49 industry models is sumarised below. Fields that are true, are marked in white, whereas false is marked in black.

Panel 1 contains some general descriptive statistics regarding the estimated HMM models. First, we observe that all models have converged to a solution given the numerical solver. Starting values in the maximum likelihood estimations are the OLS estimates. To exclude unwanted local maxima and assure the estimated solutions make sense two additional undesirable properties of the estimated Markov chains are reported:
* __Non-ergodicity/Reducability__: Either $p_{00} = 1$ or $p_{11} = 1$. This implies the existance of an absorbing state that, once reached, will be the state of nature forever onwards.
* __Periodicity__: Either $p_{00} = 0$ or $p_{11} = 0$. This implies the existance of a state that will never last for longer than one period. Whenever reached, the Markov process immediately jumps back to the other state with complete certainty. Besides the resulting perfect state-predictability, this is undesirable in an empirical exercise since it means that the estimation procedure converged to values that capture single outliers in a separate state.

Panel 2 compares the model fit of the HMM models to the corresponding values of simple CAPM models estimated by OLS. The OLS likelihood provides a lower bound for the HMM models as it can just stay at the OLS estimates if that is optimal. Given the amount of estimated parameters (eight instead of two), model selection criteria that take into account their number are better suited for comparison. In the full sample, HMM estimates still seem to add descriptive power to the standard CAPM in most industries.

Panel 3 reports the significance of the estimated parameters in OLS. The confidence level is 95%. This is reported to serve as a benchmark for the more complex models.

Panel 4 repeats the same for the HMM models. As can be seen, most alpha estimates are insignificant just as in the OLS specification. Most other estimated parameters are significant at 95%. Note that the null hypothesis for the transition probability parameters is e.g. $H_0: p_{00} = 0.5$ instead of $p_{00} = 0$.

Panel 5 repeats the significance tests for the HMM parameters against the null of the OLS estimates being true. Now most the differences arise from the idiosyncratic volatility estimates, whereas betas are frequently not distinguishable from the OLS estimates in the statistical sense.

Panel 6 reports significance tests for the estimated parameter spreads of the two regimes. The difference for each parameter pair is tested using a two-sample t-statistic, e.g. for $H_0: \beta^i_0=\beta^i_1$:
$$t_{[\beta^i_0=\beta^i_1]} = \frac{\beta^i_0-\beta^i_1}{\sqrt{\big[s.e.(\beta^i_0)\big]^2+\big[s.e.(\beta^i_1)\big]^2 - 2 \, cov(\beta^i_0,\beta^i_1)}}$$
Again, the HMM models mostly attain significant spreads in the idiosyncratic volatilities.

```python
graph_sets = [[('HMM','convergence'),('HMM','ergodicity'),('HMM','aperiodicity')],
              [('stats','likelihood_better'),('stats','aic_better'),('stats','bic_better')],
              [('OLS','alpha_significant'),('OLS','beta_significant')],
              [('HMM','alpha_0_significant'),('HMM','alpha_1_significant'),('HMM','beta_0_significant'),('HMM','beta_1_significant'),('HMM','sigma2_0_significant'),('HMM','sigma2_1_significant'),('HMM','p00_significant'),('HMM','p10_significant')],
              [('HMM','alpha_0_vs_ols_significant'),('HMM','alpha_1_vs_ols_significant'),('HMM','beta_0_vs_ols_significant'),('HMM','beta_1_vs_ols_significant'),('HMM','sigma2_0_vs_ols_significant'),('HMM','sigma2_1_vs_ols_significant')],
              [('HMM','alpha_spread_significant'),('HMM','beta_spread_significant'),('HMM','sigma2_spread_significant'),('HMM','prob_spread_significant')],
             ]

graph_ylabels = [['Convergence','Ergodicity','Aperiodicity'],
                  ['Likelihood','AIC','BIC'],
                  ['alpha','beta'],
                  ['alpha0','alpha1','beta0','beta1','sigma²_0','sigma²_1','p00','p10'],
                  ['alpha0','alpha1','beta0','beta1','sigma²_0','sigma²_1'],
                  ['alpha','beta','sigma²','p'],
                 ]

graph_titles = ['Panel 1: HMM estimation','Panel 2: HMM fit vs OLS fit (HMM fits better)','Panel 3: OLS significance (95% confidence level)','Panel 4: HMM significance (95% confidence level)','Panel 5: HMM significance against OLS estimates  (95% confidence level)','Panel 6: HMM spreads significant (95% confidence level)']

fig, axes = plt.subplots(len(graph_sets), 1, figsize=(16,12), gridspec_kw = {'height_ratios':[4,4,3,9,7,5]})

for iAx in range(0,len(graph_sets)):
    im = axes[iAx].imshow(data_full_sample[graph_sets[iAx]].T, cmap='bone')
    axes[iAx].set_xticks(np.arange(len(data_full_sample.index)))
    axes[iAx].set_yticks(np.arange(len(graph_sets[iAx])))
    axes[iAx].set_yticklabels(graph_ylabels[iAx])
    axes[iAx].set_title(graph_titles[iAx])
    axes[iAx].grid(True, linewidth=0.3, color='grey')
    if iAx is 0:
        axes[iAx].xaxis.set_tick_params(labeltop=True, labelbottom=False)
        axes[iAx].set_xticklabels(industries_monthly.columns, rotation = 'vertical', y=1.3)
    elif iAx is len(graph_sets)-1:
        axes[iAx].set_xticklabels(industries_monthly.columns, rotation = 'vertical')
    else:
        axes[iAx].xaxis.set_tick_params(labeltop=False, labelbottom=False)
```

### Estimated parameters

#### Regime switches

To describe cross-sectional differences in the characteristics of the regime-Markov-chains, we calculate expected regime durations for each estimated model as the average of both regimes' expected durations:
$$\text{Expected regime duration} = \frac{\frac{1}{1-p_{00}}+\frac{1}{1-p_{11}}}{2}$$
The chart below shows time-series averages by industry in ascending order. While most industries are estimated in an economically reasonable range, a few industries stand out with extremely short or long expected regime durations. On the lower end are mostly the industries that have been documented to be problematic in terms of fit in the previous part. The upper end is not is not necessarily problematic in the full sample, but given the estimated low switching probabilities these models likely pose a challange for estimation in shorter sample periods.

```python
barmeans = pd.DataFrame(index=industries_monthly.columns)
barmeans = data_full_sample['HMM','exp_regime_duration']
barmeans = barmeans.sort_values(ascending=True)

cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']

fig, ax = plt.subplots(1,1,figsize=(17,4))

ax.bar(range(0,49), barmeans, 0.8)
#ax.plot((data_full_sample['HMM','exp_duration_0']-data_full_sample['HMM','exp_duration_1']).abs()[barmeans.index], color='k', marker='x') #turn into vlines
ax.set_ylim(0)
ax.set_ylabel('months')
ax.set_title('Expected regime duration')
ax.set_xticks(range(0,49))
ax.set_xticklabels(barmeans.index, rotation='vertical')
plt.show()
```

#### Parameter spreads

Next, the HMM parameter estimates are compared with their OLS counterparts. Note that the Markov probabilities have no OLS equivalents and are therefore not reported here. Industries are ordered based on their beta spreads under the HMM models. The dashed black line signifies the OLS benchmark.

As can be observed from the charts, most OLS estimates are contained in the interval of HMM estimates. All estimates seem to fall in an economically reasonable range and no relation of the various spreads is obvious. Industries with relatively large spreads in one dimension do not necessarily come with large spreads in other dimensions. This hints towards different areas of regime changes being relevant for various industries.

```python
bartops = pd.DataFrame(index=industries_monthly.columns)
barbots = pd.DataFrame(index=industries_monthly.columns)
barheights = pd.DataFrame(index=industries_monthly.columns)

bartops['α'] = data_full_sample[[('HMM','alpha_0'),('HMM','alpha_1')]].max(axis=1)-data_full_sample[('OLS','alpha')]
barbots['α'] = data_full_sample[[('HMM','alpha_0'),('HMM','alpha_1')]].min(axis=1)-data_full_sample[('OLS','alpha')]
barheights['α'] = bartops['α']-barbots['α']

bartops['β'] = data_full_sample[[('HMM','beta_0'),('HMM','beta_1')]].max(axis=1)-data_full_sample[('OLS','beta')]
barbots['β'] = data_full_sample[[('HMM','beta_0'),('HMM','beta_1')]].min(axis=1)-data_full_sample[('OLS','beta')]
barheights['β'] = bartops['β']-barbots['β']

bartops['σ'] = data_full_sample[[('HMM','sigma2_0'),('HMM','sigma2_1')]].max(axis=1)**0.5-data_full_sample[('OLS','sigma2')]**0.5
barbots['σ'] = data_full_sample[[('HMM','sigma2_0'),('HMM','sigma2_1')]].min(axis=1)**0.5-data_full_sample[('OLS','sigma2')]**0.5
barheights['σ'] = bartops['σ']-barbots['σ']

# sort
bartops = barheights.sort_values('β', ascending=True)
barbots = barbots.loc[bartops.index]
barheights = barheights.loc[bartops.index]
    
cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']

fig, axes = plt.subplots(3,1,figsize=(17,9))
for iAx in range(0,3):
    ax = axes[iAx]
    ax.bar(range(0,49), barheights.iloc[:,iAx], 0.8, bottom=barbots.iloc[:,iAx], color=cycle[iAx])
    ax.axhline(0, linewidth=1.5, color='k', linestyle='--')
    ax.set_ylabel('Difference to\nOLS estimates')
    ax.set_title(bartops.columns[iAx])
    ax.set_xticks(range(0,49))
    if iAx == 0:
        ax.xaxis.set_tick_params(labeltop=True, labelbottom=False)
        ax.set_xticklabels(bartops.index, rotation='vertical', y=1.1)
    elif iAx == 2:
        ax.set_xticklabels(bartops.index, rotation='vertical')
        #ax.set_yticklabels(['','0%','10%','20%'])
    else:
        ax.set_xticklabels([])
```

### The different dimensions of regime changes

Now let us have a closer look at some HMM models to better understand what insights these models offer into the data. From the previous part we saw that the estimated regimes can differ along three dimensions: alpha, beta, and idiosyncratic risk. What follows are three selected cases where the regime changes are described in primarily one of these dimensions. Additionally, a model with questionable solution is displayed.

For each case there are three charts.
1. A scatter plot that shows how a standard CAPM model (OLS) fits the data. Observations that lay further in the past are fading in size and saturation to illustrate the time dimension.
2. A scatter plot that illustrates the 2-regime CAPM model (HMM). The existance of two regimes means that there are also two regression lines fit to the data that are illustrated as a red and a blue line. More recent observations are again plotted in larger size. In contrast to the Standard CAPM graph, observations are now color-coded according to their smoothed regime probabilities, where red points belong to regime 0 with high probability, and blue points are attributed to regime 1. Shades in between are not clearly matched.
3. A time series plot of the regime probabilities over time. One added insight of the HMM models over OLS is that the time dimension of the data is not ignored, but instead the order of occurence is accounted for in the estimation. Therefore, we obtain regime probabilities for each observation. Given the underlying markov process, regimes are clustered over time, whcih can also be observed in the HMM scatter plot. Additionally to the smoothed probabilities (with knowledge of the whole sample), filtered probabilities (only accounts for past observations) are depicted as a dashed black line.

##### Case 1: Volatility switching (Oil Industry)

The first case contains models where the regime swithing behaviour is almost exclusively decribed by two idiosyncratic volatility regimes (alpha and beta spreads are insignificant). One such example is the oil industry. Regression lines almost coincide with the OLS estimates, exept that some periods entail higher error variance.

```python
y = industries_monthly.loc[:,'Oil  '].dropna()
x = factors_monthly['Mkt-RF'].loc[y.index]

# estimate OLS CAPM model
CAPM = sm.OLS(y, sm.add_constant(x), missing='drop').fit()          
# save estimates as starting values for HMM models
a = CAPM.params['const']
b = CAPM.params['Mkt-RF']
s = CAPM.mse_resid

# estimate HMM model
init = [0.5, 0.5, a, a, 0.99*b, 1.01*b, s, s]
HMM = sm.tsa.MarkovRegression(y, k_regimes=2, exog=x, switching_trend=True, switching_exog=True, switching_variance=True, missing='drop')\
        .fit(start_params=init, cov_type='robust', method='powell')
a0 = HMM.params['const[0]']
b0 = HMM.params['x1[0]']
s0 = HMM.params['sigma2[0]']
a1 = HMM.params['const[1]']
b1 = HMM.params['x1[1]']
s1 = HMM.params['sigma2[1]']

# plot
fig,axes = plt.subplots(1,2,figsize=(17,8))

colormap = plt.cm.get_cmap('coolwarm')

ax = axes[0]
ax.axhline(0, linewidth=0.2, color='k')
ax.axvline(0, linewidth=0.2, color='k')
ax.plot(x, a+b*x, color='k', linewidth=1, label='OLS regression (α='+str(a.round(4))+', β='+str(b.round(2))+', σ='+str((s**0.5).round(4))+')')
ax.scatter(x, y, c=np.arange(0,1,1/len(x)), cmap='binary', s=np.arange(10,110,100/len(x)), marker='.', label=None, edgecolors='black', linewidths=0.25)
ax.legend(loc='lower right')
ax.set_ylabel('Industry return')
ax.set_xlabel('Market return')
ax.set_title('Standard CAPM')

ax = axes[1]
ax.axhline(0, linewidth=0.2, color='k')
ax.axvline(0, linewidth=0.2, color='k')
ax.plot(x, a0+b0*x, color=colormap(256), linewidth=1, label='Regime 0 (α='+str(a0.round(4))+', β='+str(b0.round(2))+', σ='+str((s0**0.5).round(4))+')')
ax.plot(x, a1+b1*x, color=colormap(0), linewidth=1, label='Regime 1 (α='+str(a1.round(4))+', β='+str(b1.round(2))+', σ='+str((s1**0.5).round(4))+')')
ax.scatter(x, y, c=HMM.smoothed_marginal_probabilities[0], s=np.arange(10,110,100/len(x)), marker='.', label=None, cmap=colormap, edgecolors='black', linewidths=0.25)
ax.legend(loc='lower right')
ax.set_ylabel('Industry return')
ax.set_xlabel('Market return')
ax.set_title('Hidden Markov CAPM')

plt.show()


fig,axes = plt.subplots(2,1,figsize=(17,4))

axes[0].plot(HMM.filtered_marginal_probabilities[0], label='Filtered marginal probabilities', color='k', linewidth=2, linestyle='--')
axes[0].plot(HMM.smoothed_marginal_probabilities[0], label='Smoothed marginal probabilities', color=colormap(256), linewidth=2)
plot.add_recession_bars(axes[0], startdate=y.index[0])
axes[0].set_title('Regime 0')
axes[0].set_xticklabels([])
axes[0].set_ylabel('Regime probability')
axes[0].legend()

axes[1].plot(HMM.filtered_marginal_probabilities[1], label='Filtered marginal probabilities', color='k', linewidth=2, linestyle='--')
axes[1].plot(HMM.smoothed_marginal_probabilities[1], label='Smoothed marginal probabilities', color=colormap(0), linewidth=2)
plot.add_recession_bars(axes[1], startdate=y.index[0])
axes[1].set_title('Regime 1')
axes[1].set_ylabel('Regime probability')
axes[1].legend()

plt.show()
```

## Expanding window models

Next, to investigate if HMM CAPM models add any predictive power to regular CAPM models, we use expanding windows estimation to incorporate only backward looking data. The first window uses 60 months of monthly data (with at least 48 months being available) to obtain estimates and the procedure is repeated while adding one additional observation at a time. Although the standard CAPM is often estimated in rolling windows with a shorter period, HMM estimates are more meaningful if more data is considered. That way regime shifts that only infrequently occur can be captured.

```python
window_length_monthly = 60 #minimum in the case of expanding windows
min_obs_monthly = 0.8*window_length_monthly
```

```python
industries = industries_monthly.columns
dates_monthly = industries_monthly.index[window_length_monthly-1:]
models = []
variables = []
data_out = pd.DataFrame(index=pd.MultiIndex.from_product([industries,dates_monthly],names=['Industry','Date']),\
                       columns=pd.MultiIndex.from_product([models,variables],names=['Model','Variable']))
```

```python
OLS_name = ['OLS']
HMM_model_names = ['HMM']#,'HMM fix σ exp','HMM fix α exp','HMM fix α & σ exp']
```

```python
# Add data columns to dataframe
data_out = data_out.join(pd.DataFrame(np.nan, columns=pd.MultiIndex.from_product([OLS_name, OLS.keys()]), index=data_out.index))
data_out = data_out.join(pd.DataFrame(np.nan, columns=pd.MultiIndex.from_product([['HMM full'], HMM_full.keys()]), index=data_out.index))
```

```python
%%time
# expanding window
for iWindow in range(window_length_monthly,len(industries_monthly.index)+1):

    # window selection
    month_ID = industries_monthly.index[iWindow-1]
    #include_months = industries_monthly.index[iWindow-window_length_monthly:iWindow] #rolling window
    include_months = industries_monthly.index[0:iWindow] # expanding window
    
    # load independent variable
    x = factors_monthly['Mkt-RF'].loc[include_months]

    # asset selection
    for iAsset in industries_monthly.columns:
        
        # extract industry data
        y = industries_monthly.loc[include_months, iAsset]
        
        # set minimum number of observations within window
        if y.count() < min_obs_monthly:
            continue
            
    # OLS CAPM model
        try:
            # estimate OLS CAPM model
            model = sm.OLS(y, sm.add_constant(x), missing='drop').fit()
            
            for iEntry in OLS.keys():
                data_out.loc[(iAsset, month_ID),(OLS_name, iEntry)] = eval(OLS[iEntry])
            
            # save estimates as starting values for HMM models
            a = model.params['const']
            b = model.params['Mkt-RF']
            s = model.mse_resid
            
            del model
            
        except:
            pass
    
    # HMM model 1 (full model)
        try:
            init = [0.5, 0.5, a, a, 0.99*b, 1.01*b, s, s]
            model = sm.tsa.MarkovRegression(y, k_regimes=2, exog=x, switching_trend=True, switching_exog=True, switching_variance=True, missing='drop')\
            .fit(start_params=init, cov_type='robust', method='powell')#'nm')
            
            for iEntry in HMM_full.keys():
                data_out.loc[(iAsset, month_ID),(HMM_model_names[0], iEntry)] = eval(HMM_full[iEntry])
                
            del model, init
            
        except:
            pass
        
        # clear temporary variable
        del y
    
    # print progress
    if (iWindow+1)%10 is 0:
        print('finished',iWindow-window_length_monthly+1,'out of',len(industries_monthly.index)-window_length_monthly+1)
    
    del x
```

```python
data_out['general','mean_mkt'] = factors_monthly['Mkt-RF'].mean()
```

```python
for iModel in HMM_model_names:
    # flag non-ergodic markov chains and periodic markov chains (return to a given state after 2 periods with certainty)
    data_out[iModel,'ergodicity'] = (data_out[iModel,'p00'] <= 0.9999) & (data_out[iModel,'p10'] >= 0.0001)
    data_out[iModel,'aperiodicity'] = (data_out[iModel,'p00'] >= 0.0001) & (data_out[iModel,'p10'] <= 0.9999)
    data_out[iModel,'include_model'] = (data_out[iModel,'ergodicity'] & data_out[iModel,'aperiodicity'] & data_out[iModel,'convergence'])
    
    # expected duration
    data_out[iModel,'exp_duration_0'] = 1/(1-data_out[iModel,'p00'])
    data_out[iModel,'exp_duration_1'] = 1/data_out[iModel,'p10']
    data_out[iModel,'exp_regime_duration'] = (data_out[iModel,'exp_duration_0']+data_out[iModel,'exp_duration_1'])/2
    
    # expected state probabilities (t+1)
    data_out[iModel,'exp_prob_0'] = data_out[iModel,'filt_prob_0']*data_out[iModel,'p00'] + data_out[iModel,'filt_prob_1']*data_out[iModel,'p10']
    data_out[iModel,'exp_prob_1'] = data_out[iModel,'filt_prob_0']*(1-data_out[iModel,'p00']) + data_out[iModel,'filt_prob_1']*(1-data_out[iModel,'p10'])
    
    # weighted parameters & parameter spreads & parameter t-statistics & expected values
    try:
        data_out[iModel,'weighted_alpha'] = data_out[iModel,'filt_prob_0']*data_out[iModel,'alpha_0'] + data_out[iModel,'filt_prob_1']*data_out[iModel,'alpha_1']
        data_out[iModel,'mean_weighted_alpha'] = data_out[iModel,'mean_prob_0']*data_out[iModel,'alpha_0'] + data_out[iModel,'mean_prob_1']*data_out[iModel,'alpha_1']
        data_out[iModel,'alpha_spread'] = (data_out[iModel,'alpha_0']-data_out[iModel,'alpha_1']).abs()
        data_out[iModel,'t_alpha'] = (data_out[iModel,'alpha_0']-data_out[iModel,'alpha_1']) / (data_out[iModel,'alpha_0_se']**2 + data_out[iModel,'alpha_1_se']**2 - 2*data_out[iModel,'alpha_cov'])**0.5
        data_out[iModel,'exp_alpha'] = data_out[iModel,'exp_prob_0']*data_out[iModel,'alpha_0'] + data_out[iModel,'exp_prob_1']*data_out[iModel,'alpha_1']
    except:
        pass
    
    try:
        data_out[iModel,'weighted_beta'] = data_out[iModel,'filt_prob_0']*data_out[iModel,'beta_0'] + data_out[iModel,'filt_prob_1']*data_out[iModel,'beta_1']
        data_out[iModel,'mean_weighted_beta'] = data_out[iModel,'mean_prob_0']*data_out[iModel,'beta_0'] + data_out[iModel,'mean_prob_1']*data_out[iModel,'beta_1']
        data_out[iModel,'beta_spread'] = (data_out[iModel,'beta_0']-data_out[iModel,'beta_1']).abs()
        data_out[iModel,'t_beta'] = (data_out[iModel,'beta_0']-data_out[iModel,'beta_1']) / (data_out[iModel,'beta_0_se']**2 + data_out[iModel,'beta_1_se']**2 - 2*data_out[iModel,'beta_cov'])**0.5
        data_out[iModel,'exp_beta'] = data_out[iModel,'exp_prob_0']*data_out[iModel,'beta_0'] + data_out[iModel,'exp_prob_1']*data_out[iModel,'beta_1']
    except:
        pass
    
    try:
        data_out[iModel,'weighted_sigma2'] = data_out[iModel,'filt_prob_0']*data_out[iModel,'sigma2_0'] + data_out[iModel,'filt_prob_1']*data_out[iModel,'sigma2_1']
        data_out[iModel,'mean_weighted_sigma2'] = data_out[iModel,'mean_prob_0']*data_out[iModel,'sigma2_0'] + data_out[iModel,'mean_prob_1']*data_out[iModel,'sigma2_1']
        data_out[iModel,'sigma_spread'] = (data_out[iModel,'sigma2_0']**0.5-data_out[iModel,'sigma2_1']**0.5).abs()
        data_out[iModel,'t_sigma2'] = (data_out[iModel,'sigma2_0']-data_out[iModel,'sigma2_1']) / (data_out[iModel,'sigma2_0_se']**2 + data_out[iModel,'sigma2_1_se']**2 - 2*data_out[iModel,'sigma2_cov'])**0.5
        data_out[iModel,'exp_sigma2'] = data_out[iModel,'exp_prob_0']*data_out[iModel,'sigma2_0'] + data_out[iModel,'exp_prob_1']*data_out[iModel,'sigma2_1']
    except:
        data_out[iModel,'exp_sigma2'] = data_out[iModel,'sigma2']
    
    data_out[iModel,'prob_spread'] = (data_out[iModel,'p00']+data_out[iModel,'p10']-1).abs()
    data_out[iModel,'t_prob'] = (data_out[iModel,'p00']+data_out[iModel,'p10']-1) / (data_out[iModel,'p00_se']**2 + data_out[iModel,'p10_se']**2 - 2*data_out[iModel,'prob_cov'])**0.5
    
    # means
    try:
        data_out[iModel,'mu_0'] = data_out[iModel,'alpha_0'] + data_out[iModel,'beta_0']*data_out['general','mean_mkt']
        data_out[iModel,'mu_1'] = data_out[iModel,'alpha_1'] + data_out[iModel,'beta_1']*data_out['general','mean_mkt']
    except:
        data_out[iModel,'mu_0'] = data_out[iModel,'alpha'] + data_out[iModel,'beta_0']*data_out['general','mean_mkt']
        data_out[iModel,'mu_1'] = data_out[iModel,'alpha'] + data_out[iModel,'beta_1']*data_out['general','mean_mkt']
        
    # expected means
    data_out[iModel,'exp_mean'] = data_out[iModel,'exp_prob_0']*data_out[iModel,'mu_0'] + data_out[iModel,'exp_prob_1']*data_out[iModel,'mu_1']
    
    # expected variances
    data_out[iModel,'exp_var'] = data_out[iModel,'exp_sigma2'] + \
        data_out[iModel,'exp_prob_0']*data_out[iModel,'mu_0']**2 + data_out[iModel,'exp_prob_1']*data_out[iModel,'mu_1']**2 - \
        data_out[iModel,'exp_mean']**2
    
    data_out[iModel,'markov_significant'] = ((((data_out[iModel,'p00']-0.5)/data_out[iModel,'p00_se']).abs()>1.96) & (((data_out[iModel,'p10']-0.5)/data_out[iModel,'p10_se']).abs()>1.96))
```

### Model Summary Statistics

#### Model Estimation
To summarise the models let us first look at some descriptive statistics. 
As previously, two undesirable properties of the estimated Markov chains are reported:
* __Non-ergodicity/Reducability__: Either $p_{00} = 1$ or $p_{11} = 1$. This implies the existance of an absorbing state that, once reached, will be the state of nature forever onwards.
* __Periodicity__: Either $p_{00} = 0$ or $p_{11} = 0$. This implies the existance of a state that will never last for longer than one period. Whenever reached, the Markov process immediately jumps back to the other state with complete certainty. Besides the resulting perfect state-predictability, this is undesirable in an empirical exercise since it means that the estimation procedure converged to values that capture single outliers in a separate state.
* __Insignificance__: Either $p_{00} = p_{01} = 0.5$ or $p_{11} = p_{10} = 0.5$ based on standard t-statistics with 95% confidence level. If the estimted Markov chain is not significantly different from transition probabilities of 0.5, knowledge of the last state is not informative of the next period state, and therefore the model specification is redundant.

From now onwards, __only models that have converged to ergodic, aperiodic and significant Markov chains are considered__. The large number of models makes it impossible look at each model separately, but the overall inclusion rate is sufficiently large to proceed with the analyis.

```python
convergence_table = pd.DataFrame(index=['Total models','Total converged','Convergence rate','Total ergodic','Ergodicity rate','Total aperiodic','Aperiodicity rate','Total significant','Significance rate','Total included','Inclusion rate'],\
                                columns=HMM_model_names)

for iModel in HMM_model_names:
    convergence_table.loc['Total models',iModel] = data_out[iModel,'convergence'].notna().sum()
    convergence_table.loc['Total converged',iModel] = data_out[iModel,'convergence'].sum()
    convergence_table.loc['Convergence rate',iModel] = convergence_table.loc['Total converged',iModel]/convergence_table.loc['Total models',iModel]
    
    convergence_table.loc['Total ergodic',iModel] = data_out[iModel,'ergodicity'].sum()
    convergence_table.loc['Ergodicity rate',iModel] = convergence_table.loc['Total ergodic',iModel]/convergence_table.loc['Total models',iModel]
    
    convergence_table.loc['Total aperiodic',iModel] = data_out[iModel,'aperiodicity'].sum()
    convergence_table.loc['Aperiodicity rate',iModel] = convergence_table.loc['Total aperiodic',iModel]/convergence_table.loc['Total models',iModel]
    
    convergence_table.loc['Total significant',iModel] = data_out[iModel,'markov_significant'].sum()
    convergence_table.loc['Significance rate',iModel] = convergence_table.loc['Total significant',iModel]/convergence_table.loc['Total models',iModel]
    
    convergence_table.loc['Total included',iModel] = (data_out[iModel,'include_model'] & data_out[iModel,'markov_significant']).sum()
    convergence_table.loc['Inclusion rate',iModel] = convergence_table.loc['Total included',iModel]/convergence_table.loc['Total models',iModel]
    
convergence_table
```

#### Model fit

To benchmark the regime-switching models, standard CAPM specifications with only one regime are estimated in the same data. The table below reports the frequency of cases where the Markov regime-switching models are preferred over the OLS specifications given various criteria. With eight free parameters instead of two, higher likelihood can easily be achieved. Therefore, more suitable criteria take into account the number of parameters.
Using an Akaike information criterion, the number preferred two-regime specifications slightly decreases. If instead a more restrictive Bayesian information criterion is used to select the model, in about half of the cases the Markov models are still preferred.

```python
fit_table = pd.DataFrame(index=['Likelihood','AIC','BIC'], columns=HMM_model_names)

for iModel in HMM_model_names:
    fit_table.loc['Likelihood',iModel] = (data_out[iModel,'likelihood'].loc[data_out[iModel,'likelihood'].notna()]>data_out[OLS_name[0],'likelihood'].loc[data_out[iModel,'likelihood'].notna()]).sum() / data_out[iModel,'likelihood'].notna().sum()
    fit_table.loc['AIC',iModel] = (data_out[iModel,'aic'].loc[data_out[iModel,'aic'].notna()]<data_out[OLS_name[0],'aic'].loc[data_out[iModel,'aic'].notna()]).sum() / data_out[iModel,'aic'].notna().sum()
    fit_table.loc['BIC',iModel] = (data_out[iModel,'bic'].loc[data_out[iModel,'bic'].notna()]<data_out[OLS_name[0],'bic'].loc[data_out[iModel,'bic'].notna()]).sum() / data_out[iModel,'bic'].notna().sum()

fit_table
```

If we calculate the same numbers for the included subsample, the performance of HMM models relative to the OLS CAPM looks much more favourable. Therefore, it is crucial for this analysis to select approriate estimation windows to explore the strengths of the HMM models in this context.

```python
fit_table = pd.DataFrame(index=['Likelihood','AIC','BIC'], columns=HMM_model_names)

for iModel in HMM_model_names:
    fit_table.loc['Likelihood',iModel] = (data_out[iModel,'likelihood'].loc[data_out[iModel,'include_model']]>data_out[OLS_name[0],'likelihood'].loc[data_out[iModel,'include_model']]).sum() / data_out[iModel,'include_model'].sum()
    fit_table.loc['AIC',iModel] = (data_out[iModel,'aic'].loc[data_out[iModel,'include_model']]<data_out[OLS_name[0],'aic'].loc[data_out[iModel,'include_model']]).sum() / data_out[iModel,'include_model'].sum()
    fit_table.loc['BIC',iModel] = (data_out[iModel,'bic'].loc[data_out[iModel,'include_model']]<data_out[OLS_name[0],'bic'].loc[data_out[iModel,'include_model']]).sum() / data_out[iModel,'include_model'].sum()

fit_table
```

If we compare the cross-sectional means of these statistics over time in the graph below, we recognise that there is time-series variation in how well the models match the data. To make it somewhat comparable across different sample lengths, the statistics are divided by the number of observations in the respective model (otherwise the graphs will only be strictly increasing/decreasing lines primarily driven by the number of observations). This way, the statistics have an interpretation as fit per observation. However, with an increasing number of observtions, outliers are less accurately matched and potentially drive down the fit statistics. Nevertheless, the direct comparison of different models in the same time period is still meaningful.

We observe that the variation in the fit of the Hidden Markov Models (HMM) closely matches the variation of the OLS specifications. In periods where the HMM models do a relatively poorer job coincide with periods when standard one-regime CAPM regressions also have a decresed ability to explain returns.

While model likelihood is not very informative for model comparison (OLS likelihood should be the lower bound for the HMM models), the information critera allow a direct comparison. Generally, it appears that the HMM models need to attain a critical mass of data to be consistently preferred over OLS. In particular, this point is reached in the early 70s for the AIC and around 1980 (~20 years of data) for the BIC.

#### Parameter Estimates

Next, we can again ask how different the two regimes really are, now on average over the many estimated models. For that purpose, the table below contains the spreads between the parameter estimates under both regimes. In addition, a test the statistical significance of the difference for each parameter pair is given by a two-sample t-statistic (as before), e.g. for $H_0: \beta^i_0=\beta^i_1$:
$$t_{[\beta^i_0=\beta^i_1]} = \frac{\beta^i_0-\beta^i_1}{\sqrt{\big[s.e.(\beta^i_0)\big]^2+\big[s.e.(\beta^i_1)\big]^2 - 2 \, cov(\beta^i_0,\beta^i_1)}}$$
The table reports mean and standard deviation of parameters spreads, and the percentage of test statistics being significant at the 95% level.

```python
parameter_table = pd.DataFrame(index=pd.MultiIndex.from_product([['α','β','σ','p'],['mean spread','std spread','% significant at 95%']]), columns=HMM_model_names)

for iModel in HMM_model_names:
    # alpha
    try:
        parameter_table.loc[('α','mean spread'),iModel] = data_out[iModel,'alpha_spread'].loc[data_out[iModel,'include_model']].mean()
        parameter_table.loc[('α','std spread'),iModel] = data_out[iModel,'alpha_spread'].loc[data_out[iModel,'include_model']].std()
        parameter_table.loc[('α','% significant at 95%'),iModel] = (data_out[iModel,'t_alpha'].abs()>1.96)[data_out[iModel,'include_model']].sum() / data_out[iModel,'t_alpha'].loc[data_out[iModel,'include_model']].notna().sum()
    except:
        pass
    
    # beta
    try:
        parameter_table.loc[('β','mean spread'),iModel] = data_out[iModel,'beta_spread'].loc[data_out[iModel,'include_model']].mean()
        parameter_table.loc[('β','std spread'),iModel] = data_out[iModel,'beta_spread'].loc[data_out[iModel,'include_model']].std()
        parameter_table.loc[('β','% significant at 95%'),iModel] = (data_out[iModel,'t_beta'].abs()>1.96)[data_out[iModel,'include_model']].sum() / data_out[iModel,'t_beta'].loc[data_out[iModel,'include_model']].notna().sum()
    except:
        pass
    
    # sigma
    try:
        parameter_table.loc[('σ','mean spread'),iModel] = data_out[iModel,'sigma_spread'].loc[data_out[iModel,'include_model']].mean()
        parameter_table.loc[('σ','std spread'),iModel] = data_out[iModel,'beta_spread'].loc[data_out[iModel,'include_model']].std()
        parameter_table.loc[('σ','% significant at 95%'),iModel] = (data_out[iModel,'t_sigma2'].abs()>1.96)[data_out[iModel,'include_model']].sum() / data_out[iModel,'t_sigma2'].loc[data_out[iModel,'include_model']].notna().sum()
    except:
        pass
    
    # p
    try:
        parameter_table.loc[('p','mean spread'),iModel] = data_out[iModel,'prob_spread'].loc[data_out[iModel,'include_model']].mean()
        parameter_table.loc[('p','std spread'),iModel] = data_out[iModel,'prob_spread'].loc[data_out[iModel,'include_model']].std()
        parameter_table.loc[('p','% significant at 95%'),iModel] = (data_out[iModel,'t_prob'].abs()>1.96)[data_out[iModel,'include_model']].sum() / data_out[iModel,'t_prob'].loc[data_out[iModel,'include_model']].notna().sum()
    except:
        pass

parameter_table[parameter_table.isna()] = ''
parameter_table
```

### Correlations
To further evaluate the usefulness of the Hidden Markov Models (HMM), we calculate correlations of parameter estimates of such models with their standard CAPM counterparts. In particular, we consider the following:
* __Mean weighted__: The Markov switching model parameter estimates weighted at the mean smoothed probabilities of the two regimes over the observation window.
* __Last period weighted__: The Markov switching model estimates weighted at the filtered regime-probabilities in the last month in the observation window.
* __Expected parameter__: The Markov switching model estimates weighted at the filtered regime-probabilities in the last month in the observation window, iterated forward by one month through the transition matrix.
* __Same period CAPM__: OLS estimates from CAPM regressions in the same observation window as in the HMM case.
* __Realised CAPM__: OLS estimates from CAPM regressions using 1 month of daily data.

Below there are six comparisons reported in the table (correlations):
* __Mean HMM vs. CAPM__: The first sanity check for the HMM models is to compare their estimates with the one-regime case of the standard CAPM. To have a sensible comparison, regime-parameters are weighted at the mean probabilites over the respective sample. Estimates obtained in such a way are very stongly correlated to their CAPM equivalents.
* __Last period weighted HMM at t vs. Realised at t__: Next, we ask if the most recent monthly estimates for each model are correlated with actually realised values. For that purpose, we correlate probability-weighted estimates at the end of each sample with realised CAPM estimates from daily data within the respective month. Correlations in this comparison indicate that HMM models do reasonably well in describing parameters realised within a given month. As a benchmark, the OLS CAPM is additionally correlated to the realised regressions, providing weaker correlations. Note that this is a contemperaneous relation, not a predictive one.
* __Expected HMM parameter at t-1 vs. Realised at t__: We then examine the predictive ability of HMM models with regards to realised volatility in the subsequent month. Instead of using static regime-probabilities, we can iterate one period forward by using the transition matrix. Such predicted regime probabilities are obtained purely from historical data and can therefore serve as a 1-month forecast. Compared to the previous contemperaneous relation, correlations are much smaller. Especially puzzling is the non-existence of predictive power with regards to idiosyncratic volatility, especially when benchmarked against OLS estimates.
* __OLS CAPM at t vs. Realised at t__: The expanding window estimates are modereately correlated to the realised values in the final month. 
* __OLS CAPM at t vs. Realised at t+1__: When looking at the predictive relation, not much changes as compared to the contemperaneous relation as expanding window OLS estimates are almost perfectly autocorreled.
* __Realised at t vs. Realised at t+1__: To complete the list of benchmarks, the autocorrelations of realised regression estimates are also reported.

The table contains Pearson correlation coefficients calculated over all time periods and assets given the above described sample restrictons.

### Comparison of model forecast errors

To assess the predictive ability of the HMM model along several dimensions, 1-month expectations are calculated for the three CAPM specifications and compared to their realisations in the following month. Realised total volatility is directly determined from daily returns, while alpha, beta and idiosyncratic volatility are estimated in intra-month regressions as previously.

In order to arrive at model predictions for these dimensions, expected model parameters need to be defined (alpha, beta, and idiosyncratic volatility). In the case of OLS and Real OLS, next-month expected parameters correspond to the previous month's estimates. For the HMM models, expectations are formed by iterating one month forward through the Markov transition matrix.

Model-implied expectations of return and total volatility are then given by:
$$\mathbb{E}_t\big[r^i_{t+1}\big] = \mathbb{E}_t\big[\beta^i_{t+1}\big] \times \mathbb{E}_t\big[r^M_{t+1}\big]$$

$$\mathbb{E}_t\big[\sigma^{i, total}_{t+1}\big] = \sqrt{\mathbb{E}_t\big[\beta^i_{t+1}\big]^2 \times \mathbb{E}_t\big[(\sigma^{M}_{t+1})^2\big] + \mathbb{E}_t\big[(\sigma^{i, idiosyncratic}_{t+1})^2\big]}$$

Where $\mathbb{E}_t\big[r^M_{t+1}\big]$ and $\mathbb{E}_t\big[(\sigma^{M}_{t+1})^2\big]$ are the long-time averages of market return and market realised volatility.

```python
industries_real_var = industries_daily.groupby(pd.Grouper(freq='M')).var(ddof=0)*22
industries_real_var.index = [pd.datetime(x.year, x.month, 1) for x in industries_real_var.index.tolist()]
```

```python
market_mean = factors_monthly['Mkt-RF'].mean()
market_var = (factors_daily['Mkt-RF'].groupby(pd.Grouper(freq='M')).var(ddof=0)*22).mean()
```

```python
forecast_data = data_out[[('Realised CAPM','alpha'),('Realised CAPM','beta'),('Realised CAPM','sigma2')]]
forecast_data['Realised CAPM','alpha'] = forecast_data['Realised CAPM','alpha']*22
forecast_data['Realised CAPM','sigma'] = (forecast_data['Realised CAPM','sigma2']*22)**0.5
forecast_data.loc[:,('Returns','return')] = [industries_monthly.loc[indu,day] for (day,indu) in forecast_data.index]
forecast_data.loc[:,('Returns','volatility')] = [industries_real_var.loc[indu,day]**0.5 for (day,indu) in forecast_data.index]

forecast_data = forecast_data.loc[data_out['HMM','incl_crit']]

forecast_data['OLS','exp_ret'] = data_out['OLS','beta']*market_mean #+data_out['OLS','alpha'] 
forecast_data['OLS','exp_vola'] = (data_out['OLS','beta']**2*market_var+data_out['OLS','sigma2'])**0.5
forecast_data['OLS','exp_idio_vola'] = data_out['OLS','sigma2']**0.5
forecast_data['OLS','exp_alpha'] = data_out['OLS','alpha']
forecast_data['OLS','exp_beta'] = data_out['OLS','beta']

forecast_data['Real OLS','exp_ret'] = data_out['Realised CAPM','beta']*market_mean #+data_out['Realised CAPM','alpha']
forecast_data['Real OLS','exp_vola'] = (data_out['Realised CAPM','beta']**2*market_var+data_out['Realised CAPM','sigma2']*22)**0.5
forecast_data['Real OLS','exp_idio_vola'] = (data_out['Realised CAPM','sigma2']*22)**0.5
forecast_data['Real OLS','exp_alpha'] = data_out['Realised CAPM','alpha']*22
forecast_data['Real OLS','exp_beta'] = data_out['Realised CAPM','beta']

forecast_data['HMM','exp_ret'] = data_out['HMM','exp_beta']*market_mean #+data_out['HMM','exp_alpha']
forecast_data['HMM','exp_vola'] = (data_out['HMM','exp_beta']**2*market_var+data_out['HMM','exp_sigma2'])**0.5
forecast_data['HMM','exp_idio_vola'] = data_out['HMM','exp_sigma2']**0.5
forecast_data['HMM','exp_alpha'] = data_out['HMM','exp_alpha']
forecast_data['HMM','exp_beta'] = data_out['HMM','exp_beta']
```

```python
forecast_table = pd.DataFrame(index=['return','total volatility','alpha','beta','idiosyncratic volatility'], columns=pd.MultiIndex.from_product([['MSE','MAD','MAPE'],['OLS','HMM','Real OLS']], names=['Loss function','Model']))

for iModel in ['OLS','HMM','Real OLS']:
    forecast_table.loc['return',('MSE',iModel)] = ((forecast_data['Returns','return']-forecast_data[iModel,'exp_ret'].groupby('Industry').shift(1))**2).mean()
    forecast_table.loc['total volatility',('MSE',iModel)] = ((forecast_data['Returns','volatility']-forecast_data[iModel,'exp_vola'].groupby('Industry').shift(1))**2).mean()
    forecast_table.loc['alpha',('MSE',iModel)] = ((forecast_data['Realised CAPM','alpha']-forecast_data[iModel,'exp_alpha'].groupby('Industry').shift(1))**2).mean()
    forecast_table.loc['beta',('MSE',iModel)] = ((forecast_data['Realised CAPM','beta']-forecast_data[iModel,'exp_beta'].groupby('Industry').shift(1))**2).mean()
    forecast_table.loc['idiosyncratic volatility',('MSE',iModel)] = ((forecast_data['Realised CAPM','sigma']-forecast_data[iModel,'exp_idio_vola'].groupby('Industry').shift(1))**2).mean()
    
    forecast_table.loc['return',('MAD',iModel)] = (forecast_data['Returns','return']-forecast_data[iModel,'exp_ret'].groupby('Industry').shift(1)).abs().mean()
    forecast_table.loc['total volatility',('MAD',iModel)] = (forecast_data['Returns','volatility']-forecast_data[iModel,'exp_vola'].groupby('Industry').shift(1)).abs().mean()
    forecast_table.loc['alpha',('MAD',iModel)] = (forecast_data['Realised CAPM','alpha']-forecast_data[iModel,'exp_alpha'].groupby('Industry').shift(1)).abs().mean()
    forecast_table.loc['beta',('MAD',iModel)] = (forecast_data['Realised CAPM','beta']-forecast_data[iModel,'exp_beta'].groupby('Industry').shift(1)).abs().mean()
    forecast_table.loc['idiosyncratic volatility',('MAD',iModel)] = (forecast_data['Realised CAPM','sigma']-forecast_data[iModel,'exp_idio_vola'].groupby('Industry').shift(1)).abs().mean()
    
    forecast_table.loc['return',('MAPE',iModel)] = (1-forecast_data['Returns','return']/forecast_data[iModel,'exp_ret'].groupby('Industry').shift(1)).abs().mean()
    forecast_table.loc['total volatility',('MAPE',iModel)] = (1-forecast_data['Returns','volatility']/forecast_data[iModel,'exp_vola'].groupby('Industry').shift(1)).abs().mean()
    forecast_table.loc['alpha',('MAPE',iModel)] = (1-forecast_data['Realised CAPM','alpha']/forecast_data[iModel,'exp_alpha'].groupby('Industry').shift(1)).abs().mean()
    forecast_table.loc['beta',('MAPE',iModel)] = (forecast_data['Realised CAPM','beta']/forecast_data[iModel,'exp_beta'].groupby('Industry').shift(1)).abs().mean()
    forecast_table.loc['idiosyncratic volatility',('MAPE',iModel)] = (1-forecast_data['Realised CAPM','sigma']/forecast_data[iModel,'exp_idio_vola'].groupby('Industry').shift(1)).abs().mean()
```

Compared to the OLS benchmark, the HMM model performs better in predicting total volatility, beta, and idiosyncrativ volatility. Returns and alphas are forecasted with the almost same (poor) precision. Given that the OLS specification is estimated in the same data as the HMM model, this is the appropriate benchmark. Nevertheless, the forecast errors are also assessed in comparison to the Realised OLS regressions, that are estimated from daily data. The realised OLS model offers an improvement over both monthly models.

Predictive performance is assessed based on mean squared error, mean absolute deviation and mean absolute percentage error. Given the many realisation close to zero, the latter has only limited conclusiveness.

### Predictive panel regressions

```python
reg_data = forecast_data
```

Next, we run predictive regressions to estimate the relation of model predictions of the different models with next-period realisations. Predictive panel regressions of the following form are estimated:
$$X^i_{t+1} = (\gamma_0) + \gamma_1 \mathbb{E}_t\big[X^i_{t+1}\big] + (\tau_t) + (\eta^i) + \epsilon^i_{t+1}$$
$X^i_{t}$ are the realisations of the respective variable and $\mathbb{E}_t\big[X^i_{t+1}\big]$ is the previous-period prediction under the respective model. $\tau_t$ and $\eta^i$ are time- and industry fixed effects.

Reported in the table below are the estimates and t-statistics (Driscoll-Kraay) from such regressions. Additionally, the R² of such regressions is reported both for the regressors only, and for the regressors together with the fixed effects.

Note that there are some potential problems with these specifications:
* Estimation through static panel data techniques (standard fixed effects estimation) in a dynamic panel is biased if lags of the dependent variable are included (as is the case for the Real OLS specifications), and right hand side variables should be instrumented appropriately (e.g. Anderson-Hsiao IV or Arellano-Bond GMM estimator).
* The time series of volatility, alpha, beta can be non-stationary and correlations may be spurious.
* The 'together' specifications include multicollinear independent variables that affect parameter estimation.

Generally, models that achieve conditional unbiasedness are preferred, i.e. the intercept in the pooled regressions without fixed effects are desired be zero and the regression coefficient for the model' expectation close to unit.

#### Returns

Generally, the level of return predictability is very little to none. The only exception to this is the expected return from the standard CAPM OLS regressions. which curiously significantly predict returns with a negative sign. Given the expanding window estimation of these models, there is very little time-series variation in betas, and hence expected returns. Why this little variation predicts returns nevertheless with a negative sign is a puzzle similar to the low-beta anomaly.

```python
var_names = ['Real OLS','OLS','HMM']#,'HMM weighted']
param_names = ['Real OLS','OLS','HMM','together']
pred_table = pd.DataFrame(index=pd.MultiIndex.from_product([['constant']+var_names,['est.','t-stat']]), columns=pd.MultiIndex.from_product([param_names,np.arange(1,5)], names=['Predictor','']))

for dep in param_names:
    col = 0
    for (tim,ent) in [(False,False),(True,False),(False,True),(True,True)]:
        col += 1
        try:
            x_data = reg_data[(dep,'exp_ret')].to_frame()
        except:
            x_data = reg_data[[('Real OLS','exp_ret'),('OLS','exp_ret'),('HMM','exp_ret')]]
        
        if tim or ent is True:
            # run regression
            reg = lm.PanelOLS(reg_data['Returns','return'], x_data.groupby('Industry').shift(1), entity_effects=ent, time_effects=tim).fit(cov_type='kernel')
            # store
            pred_table.loc[[(x,'est.') for x in x_data.columns.get_level_values(0)],(dep,col)] = reg.params.values
            pred_table.loc[[(x,'t-stat') for x in x_data.columns.get_level_values(0)],(dep,col)] = reg.tstats.values
        else:
            # run regression
            reg = lm.PanelOLS(reg_data['Returns','return'], sm.add_constant(x_data.groupby('Industry').shift(1)), entity_effects=ent, time_effects=tim).fit(cov_type='kernel')
            #,('HMM','weighted_'+dep)
            # store
            pred_table.loc[[(x,'est.') for x in ['constant']+x_data.columns.get_level_values(0).tolist()],(dep,col)] = reg.params.values
            pred_table.loc[[(x,'t-stat') for x in ['constant']+x_data.columns.get_level_values(0).tolist()],(dep,col)] = reg.tstats.values
            
        pred_table.loc[('fixed effects','time'),(dep,col)] = tim
        pred_table.loc[('fixed effects','industry'),(dep,col)] = ent
        pred_table.loc[('stats','R² (model)'),(dep,col)] = reg.rsquared
        #pred_table.loc[('stats','R² (overall)'),(dep,col)] = reg.rsquared_overall
        #pred_table.loc[('stats','R² (within)'),(dep,col)] = reg.rsquared_within
        #pred_table.loc[('stats','R² (between)'),(dep,col)] = reg.rsquared_between
        pred_table.loc[('stats','R² (inclusive)'),(dep,col)] = reg.rsquared_inclusive
        pred_table.loc[('stats','N'),(dep,col)] = reg.nobs
        
pred_table[pred_table.isna()] = ''
pred_table[pred_table == False] = ''
```

## Ambiguity measurement

```python
data_amb = pd.DataFrame(index=data_out.index)
```

Given the two-regime process, a measure of (expected) ambiguity in such a setting chould take into account:
* The (second-order) probabilities of being in either regime
* The distance between the two regime prior distributions
* The dispersion of parameters under the two regimes

To get a feel for the range of values the calculated measures take, we show plots for cross-sectional statistics over time.

### Regime-Probabilities
Regarding the probabilities, we can obtain a proxy of ambiguity by considering the dispersion of regime probabilities. If we assume that the agents' perceived ambiguity is related to the uncertainty of being in either of the two regimes, we can construct a measure of ambiguity based on the estimated next-period probabilities of being in either of the states. In this framework, maximum ambiguity is reached when the two states are equally likely (p=0.5). On the other hand there is no ambiguity if all the probability mass is on one state (either p=0 or p=1). Therefore, measures could be:
* The L2-distance to the point of equal probability, scaled between 0 and 1:
$$\mathbb{D}_t\big[r^i_{t+1}\big] = 1-4 \, \big(P_t[S^i_{t+1} = 0]-0.5\big)^2$$
* Shannon entropy:
$$\mathbb{H}_t\big[r^i_{t+1}\big] = -\sum^1_{s_{t+1}=0} P_t[S^i_{t+1} = s_{t+1}] * log_2\big(P_t[S^i_{t+1} = s_{t+1}]\big)$$

Both measures take a value of one if both states are equally likely (maximum ambiguity), and zero if there is no ambiguity about the next period return distribution. Hence they take very similar values.

```python
data_amb['prob_distance'] = 1-4*(data_out['HMM','exp_prob_0']-0.5)**2
data_amb['prob_entropy'] = -(data_out['HMM','exp_prob_0']*np.log2(data_out['HMM','exp_prob_0']) + data_out['HMM','exp_prob_1']*np.log2(data_out['HMM','exp_prob_1'].loc[data_out['HMM','exp_prob_1']>0]))
data_amb = data_amb.loc[data_out['HMM','incl_crit']]
```

### Regime-Dispersion
Regarding the distance between the two priors, we can calculate a distance measure between probability distributions, e.g:
* Kolmogorov-Smirnov distance:
$$\mathbb{KS}_t\big[r^i_{t+1}\big] = \sup_{r^i_{t+1}}\Big|\Phi_0(r^i_{t+1})-\Phi_1(r^i_{t+1})\Big| \quad \text{, where $\Phi(.)$ denotes the normal cumulative density function}$$
* Mahalanobis distance (Bregman divergence):
$$\mathbb{M}_t\big[r^i_{t+1}\big] = \sqrt{\frac{(\mu^i_0-\mu^i_1)^2}{\frac{(\sigma^i_0)^2+(\sigma^i_1)^2}{2}}}$$

In order to parametrise the probability distributions (i.e. $\mu^i_{s_{t+1}}$ and $\sigma^i_{s_{t+1}}$), the previous estimates are used in combination with the full-sample average of the excess return on the market portfolio (as $\mathbb{E}_t\big[r^M_{t+1}\big]$).

```python
%%time
data_amb['mahalanobis_distance'] = ((data_out['HMM','mu_0']-data_out['HMM','mu_1'])**2 / ((data_out['HMM','sigma2_0']+data_out['HMM','sigma2_1'])/2))**0.5
    
data_amb['kolmogorov_smirnov'] = np.nan
for iAsset in industries_monthly.columns:
    for iMonth in data_out.index.get_level_values(1).unique():
        mu_0 = data_out.loc[(iAsset,iMonth),('HMM','mu_0')]
        mu_1 = data_out.loc[(iAsset,iMonth),('HMM','mu_1')]
        try:
            sigma2_0 = data_out.loc[(iAsset,iMonth),('HMM','sigma2_0')]
            sigma2_1 = data_out.loc[(iAsset,iMonth),('HMM','sigma2_0')]
        except:
            sigma2_0 = data_out.loc[(iAsset,iMonth),('HMM','sigma2')]
            sigma2_1 = data_out.loc[(iAsset,iMonth),('HMM','sigma2')]
            
        data_amb.loc[(iAsset,iMonth),'kolmogorov_smirnov'] = calc.calculate_kolmogorov_smirnov_distance(mu_0, mu_1, sigma2_0, sigma2_1)
        
    print(iAsset)

data_amb = data_amb.loc[data_out['HMM','incl_crit']]
```

### Pooled Regressions

#### Predicting Returns

To establish whether the ambiguity measure predicts returns on the subsequent period, we run pooled regressions of the following form:
$$ r_{t+1}^i = \theta_0 + \theta_1 \, \text{Expected Ambiguity}_t\big[r^i_{t+1}\big] + \gamma \, X^i_t + \varepsilon_{t+1}^i $$

$\text{Expected Ambiguity}_t\big[r^i_{t+1}\big]$ is one of the ambiguity measures defined in the above section and is the independent variable of interest.

$X^i_t$ is a set of conditioning variables. After considering univariate regressions of returns on ambiguity as depicted in the scatter plot below, we include the expected first two moments of the mixed prior distribution to contrast the effect of ambiguity with that of risk. These expected moments incorporate the expected parameters obtained from the Markov chain of regime probabilities. In another specification we can also directly include the expected parameters $\alpha$, $\beta$ and $\sigma$ under the HMM model. Last, we include the lagged realised parameters from the previous months' daily observations.

To disentangle the cross-sectional effects from the time-series, we additionally allow for time-fixed-effects $\lambda_t$ and industry-fixed-effects $\alpha^i$ in alternative specifications:
$$ r_{t+1}^i = \theta_1 \, \text{Expected Ambiguity}_t\big[r^i_{t+1}\big] + \gamma \, X^i_t + \alpha^i + \lambda_t +\varepsilon_{t+1}^i $$

T-statistics are obtained from Driscoll-Kraay standard errors.

```python
dependents = [data_full['Returns'].columns[0]]

ind_names = {'c': 'constant',
              'S': '$\mathbb{S}_t[r^i_{t+1}]$',
              'KS': '$\mathbb{KS}_t[r^i_{t+1}]$',
              'E': '$\mathbb{E}_t[r^i_{t+1}]$',
              'V': '$\mathbb{V}_t[r^i_{t+1}]$',
              'ea': '$\mathbb{E}_t[α^i_{t+1}]$',
              'eb': '$\mathbb{E}_t[β^i_{t+1}]$',
              'es': '$\mathbb{E}_t[(σ_{idio})^i_{t+1}]$',
              'ra': 'realised $α_{t-1}$',
              'rb': 'realised $β_{t-1}$',
              'rs': 'realised $(σ_{idio})^i_{t-1}$',
             }

ind_vars = {'c': ('Ones','constant'),
                'S': ('ambiguity','prob_entropy'),
                'KS': ('ambiguity','kolmogorov_smirnov'),
                'E': ('HMM','exp_ret'),
                'V': ('HMM','exp_vola'),
                'ea': ('HMM','exp_alpha'),
                'eb': ('HMM','exp_beta'),
                'es': ('HMM','exp_idio_vola'),
                'ra': ('Real OLS','exp_alpha'),
                'rb': ('Real OLS','exp_beta'),
                'rs': ('Real OLS','exp_idio_vola'),
               }

independents = ['S','KS']

controls = [[],
            ['E','V'],
            ['ea','eb','es'],
            ['ra','rb','rs'],
            ]

reg_table = pd.DataFrame(index=pd.MultiIndex.from_product([list(ind_names.values()),['est.','t-stat']]).append(pd.MultiIndex.from_product([['fixed effect'],['industry','time']])).append(pd.MultiIndex.from_product([['stats'],['R²','R² (inclusive)','N']])),\
                        columns=pd.MultiIndex.from_product([dependents,['('+str(x)+')' for x in np.arange(1,len(independents)*(len(controls)+3)+1)]]))
```

```python
count_all = 0

for dep in dependents:
    count = 0
    for ind in independents:
        for cont in controls:
            
            count += 1
            count_all += 1
                    
            # run regression
            reg = lm.PanelOLS(data_full[('Returns',dep)], data_full[[ind_vars['c']]+[ind_vars[ind]]+[ind_vars[x] for x in cont]].shift(1), entity_effects=False, time_effects=False).fit(cov_type='kernel')
                        
            # store
            reg_table.loc[([ind_names['c']]+[ind_names[ind]]+[ind_names[x] for x in cont],'est.'),(dep,'('+str(count)+')')] = reg.params.values
            # reg_table.loc[([feat_names['c']]+[feat_names[ind]]+[feat_names[x] for x in cont],'s.e.'),(iModel,iLabel,'('+str(count)+')')] = reg.std_errors.values
            reg_table.loc[([ind_names['c']]+[ind_names[ind]]+[ind_names[x] for x in cont],'t-stat'),(dep,'('+str(count)+')')] = reg.tstats.values
            reg_table.loc[('stats','R²'),(dep,'('+str(count)+')')] = reg.rsquared
            reg_table.loc[('stats','R² (inclusive)'),(dep,'('+str(count)+')')] = reg.rsquared_inclusive
            reg_table.loc[('stats','N'),(dep,'('+str(count)+')')] = reg.nobs

        for (ent,tim) in [(True,False),(False,True),(True,True)]:
            
            count += 1
            count_all += 1
            
            # run regressions with fixed effects
            reg = lm.PanelOLS(data_full[('Returns',dep)], data_full[[ind_vars[ind]]].shift(1), entity_effects=ent, time_effects=tim).fit(cov_type='kernel')
                        
            # store
            reg_table.loc[([ind_names[ind]],'est.'),(dep,'('+str(count)+')')] = reg.params.values
            # reg_table.loc[([feat_names['c']]+[feat_names[ind]]+[feat_names[x] for x in cont],'s.e.'),(iModel,iLabel,'('+str(count)+')')] = reg.std_errors.values
            reg_table.loc[([ind_names[ind]],'t-stat'),(dep,'('+str(count)+')')] = reg.tstats.values
            reg_table.loc[('fixed effect','industry'),(dep,'('+str(count)+')')] = ent
            reg_table.loc[('fixed effect','time'),(dep,'('+str(count)+')')] = tim
            reg_table.loc[('stats','R²'),(dep,'('+str(count)+')')] = reg.rsquared
            reg_table.loc[('stats','R² (inclusive)'),(dep,'('+str(count)+')')] = reg.rsquared_inclusive
            reg_table.loc[('stats','N'),(dep,'('+str(count)+')')] = reg.nobs   
```

### Portfolio Sorts

Another way of testing the ambiguity variables' predictive abilities is through portfolio sorts. We employ univariate sorts and bivariate sorts together with risk measures. Bivariate sorts are carried out both simultaneously and sequentially.

#### Functions

```python
def sort_portfolios(returns, ranking_variable, n_portfolios, lags=1, return_assets=False):
    # align periods
    sorting_variable = ranking_variable.shift(lags)
    
    # set up parameters
    [t,n] = returns.shape
    include = returns.notna() & sorting_variable.notna()
    n_period = include.sum(axis=1)

    # sort assets
    returns[~include] = np.nan
    sorting_variable[~include] = np.nan
    cutoff_ranks = np.dot(n_period.values.reshape(t,1)/n_portfolios,np.arange(n_portfolios+1).reshape(1,n_portfolios+1)).round()
    asset_ranks = sorting_variable.rank(axis=1)
    
    # set up output frames
    portfolio_returns = pd.DataFrame(index=returns.index,columns=range(1,n_portfolios+1))
    portfolio_assets = pd.DataFrame(index=returns.index,columns=range(1,n_portfolios+1))
    portfolio_mapping = pd.DataFrame(index=returns.index, columns=returns.columns)
    
    # calculate outputs
    for i_portfolio in range(0,n_portfolios):
        lower = cutoff_ranks[:,i_portfolio].reshape(t,1).repeat(n, axis=1)
        upper = cutoff_ranks[:,i_portfolio+1].reshape(t,1).repeat(n, axis=1)
        portfolio_returns[i_portfolio+1] = returns[(asset_ranks>lower) & (asset_ranks<=upper)].mean(axis=1)
        portfolio_assets[i_portfolio+1] = ((asset_ranks>lower) & (asset_ranks<=upper)).sum(axis=1)
        portfolio_mapping[(asset_ranks>lower) & (asset_ranks<=upper)] = i_portfolio
    
    # outputs
    if return_assets == False:
        return portfolio_returns
    else:
        return portfolio_returns, portfolio_assets, portfolio_mapping
```

```python
def double_sort_portfolios(returns, ranking_variable_1, ranking_variable_2, n_portfolios_1, n_portfolios_2, lags_1=1, lags_2=1, return_assets=False):
    # identify missing values
    exclude = returns.isna() | ranking_variable_1.shift(lags_1).isna() | ranking_variable_2.shift(lags_2).isna()
    returns[exclude] = np.nan
    
    # first sort
    portfolio_mapping_1 = sort_portfolios(returns, ranking_variable_1, n_portfolios_1, lags_1, return_assets=True)[2]
    
    # second sorts
    portfolio_mapping_2 = pd.DataFrame(0, index=portfolio_mapping_1.index, columns=portfolio_mapping_1.columns)
    for i_portfolio_2 in range(0,n_portfolios_2):
        subportfolio_returns = returns[portfolio_mapping_1 == i_portfolio_2]
        portfolio_mapping_2 += (sort_portfolios(subportfolio_returns, ranking_variable_2, n_portfolios_2, lags_2, return_assets=True)[2]).fillna(0)
    portfolio_mapping_2[exclude] = np.nan
    
    # combined sort
    portfolio_mapping = portfolio_mapping_1*n_portfolios_1 + portfolio_mapping_2
    
    # set up output frames
    portfolio_returns = pd.DataFrame(index=returns.index,columns=[str(i_portfolio_1+1)+','+str(i_portfolio_2+1) for i_portfolio_1 in range(0,n_portfolios_1) for i_portfolio_2 in range(0,n_portfolios_2)])
    portfolio_assets = pd.DataFrame(index=returns.index,columns=[str(i_portfolio_1+1)+','+str(i_portfolio_2+1) for i_portfolio_1 in range(0,n_portfolios_1) for i_portfolio_2 in range(0,n_portfolios_2)])
    
    # calculate outputs
    for i_portfolio_all in range(0,n_portfolios_1*n_portfolios_2):
        portfolio_returns.iloc[:,i_portfolio_all] = returns[portfolio_mapping == i_portfolio_all].mean(axis=1)
        portfolio_assets.iloc[:,i_portfolio_all] = (portfolio_mapping == i_portfolio_all).sum(axis=1)
        
    # outputs
    if return_assets == False:
        return portfolio_returns
    else:
        return portfolio_returns, portfolio_assets, portfolio_mapping
```

```python
def double_sort_portfolios_simultaneously(returns, ranking_variable_1, ranking_variable_2, n_portfolios_1, n_portfolios_2, lags_1=1, lags_2=1, return_assets=False):
    # identify missing values
    exclude = returns.isna() | ranking_variable_1.shift(lags_1).isna() | ranking_variable_2.shift(lags_2).isna()
    returns[exclude] = np.nan
    
    # first sort
    portfolio_mapping_1 = sort_portfolios(returns, ranking_variable_1, n_portfolios_1, lags_1, return_assets=True)[2]
    
    # second sorts
    portfolio_mapping_2 = sort_portfolios(returns, ranking_variable_2, n_portfolios_2, lags_2, return_assets=True)[2]
    
    # combined sort
    portfolio_mapping = portfolio_mapping_1*n_portfolios_1 + portfolio_mapping_2
    
    # set up output frames
    portfolio_returns = pd.DataFrame(index=returns.index,columns=[str(i_portfolio_1+1)+','+str(i_portfolio_2+1) for i_portfolio_1 in range(0,n_portfolios_1) for i_portfolio_2 in range(0,n_portfolios_2)])
    portfolio_assets = pd.DataFrame(index=returns.index,columns=[str(i_portfolio_1+1)+','+str(i_portfolio_2+1) for i_portfolio_1 in range(0,n_portfolios_1) for i_portfolio_2 in range(0,n_portfolios_2)])
    
    # calculate outputs
    for i_portfolio_all in range(0,n_portfolios_1*n_portfolios_2):
        portfolio_returns.iloc[:,i_portfolio_all] = returns[portfolio_mapping == i_portfolio_all].mean(axis=1)
        portfolio_assets.iloc[:,i_portfolio_all] = (portfolio_mapping == i_portfolio_all).sum(axis=1)
        
    # outputs
    if return_assets == False:
        return portfolio_returns
    else:
        return portfolio_returns, portfolio_assets, portfolio_mapping
```

### Does ambiguity add power to predictions?

With the Hidden Markov models making explicit predictions about next period expected returns based on the time-varying beta loadings, we can ask the question if next period return prediction can be improved by adding one of the ambiguity variables as predictors. For that purpose, the baseline case without ambiguity variable is ran first and contrasted with enhanced specifications. Specifically, the regressions take the following form:
$$ r_{t+1}^i = (\theta_0) + \theta_1 \, \mathbb{E}_t\big[r^i_{t+1}\big] + \theta_2 \, \text{Expected Ambiguity}_t\big[r^i_{t+1}\big] + \gamma \, X^i_t + (\alpha^i) + (\lambda_t) + \varepsilon_{t+1}^i $$

```python
var_names = ['Base case','Entropy','Kolmogorov-Smirnov']
amb_names = ['HMM','prob_entropy','kolmogorov_smirnov']

pred_table = pd.DataFrame(index=pd.MultiIndex.from_product([['constant']+amb_names,['est.','t-stat']]), columns=pd.MultiIndex.from_product([var_names,np.arange(1,5)], names=['Predictor','']))

for (amb,var) in zip(amb_names,var_names):
    col = 0
    for (tim,ent) in [(False,False),(True,False),(False,True),(True,True)]:
        col += 1
        if amb is 'HMM':
            x_data = data_full[('HMM','exp_ret')].to_frame()
        else:
            x_data = data_full[[('HMM','exp_ret'),('ambiguity',amb)]]
        
        if tim or ent is True:
            # run regression
            reg = lm.PanelOLS(data_full['Returns','return'], x_data.groupby('Industry').shift(1), entity_effects=ent, time_effects=tim).fit(cov_type='kernel')
            # store
            pred_table.loc[[(x,'est.') for x in ['HMM']+list(x_data.columns.get_level_values(1)[1:])],(var,col)] = reg.params.values
            pred_table.loc[[(x,'t-stat') for x in ['HMM']+list(x_data.columns.get_level_values(1)[1:])],(var,col)] = reg.tstats.values
        else:
            # run regression
            reg = lm.PanelOLS(data_full['Returns','return'], sm.add_constant(x_data.groupby('Industry').shift(1)), entity_effects=ent, time_effects=tim).fit(cov_type='kernel')
            # store
            pred_table.loc[[(x,'est.') for x in ['constant','HMM']+list(x_data.columns.get_level_values(1)[1:])],(var,col)] = reg.params.values
            pred_table.loc[[(x,'t-stat') for x in ['constant','HMM']+list(x_data.columns.get_level_values(1)[1:])],(var,col)] = reg.tstats.values
            
        pred_table.loc[('fixed effects','time'),(var,col)] = tim
        pred_table.loc[('fixed effects','industry'),(var,col)] = ent
        pred_table.loc[('stats','R² (model)'),(var,col)] = reg.rsquared
        #pred_table.loc[('stats','R² (overall)'),(dep,col)] = reg.rsquared_overall
        #pred_table.loc[('stats','R² (within)'),(dep,col)] = reg.rsquared_within
        #pred_table.loc[('stats','R² (between)'),(dep,col)] = reg.rsquared_between
        pred_table.loc[('stats','R² (inclusive)'),(var,col)] = reg.rsquared_inclusive
        pred_table.loc[('stats','N'),(var,col)] = reg.nobs
        
pred_table[pred_table.isna()] = ''
pred_table[pred_table == False] = ''
```

```python
pred_table.replace(np.nan,'').replace(False,'')
```

Although the R² slightly increases when ambiguity variables are added, the changes are too small to draw conclusions. More, the coefficients on the ambiguity variables are insignificant. However, they mostly seem to have a much stronger relation to subsequent returns than the model expected returns themselves.

### Is cross-sectional ambiguity a priced factor?

To investigate this question, we treat the ambiguity sorted portfolios as risk factors and run Fama-MacBeth regressions from 60-month three-factor models together with the market factor. That is, for each industry in each month, we estimate three-factor loadings over a rolling window of 60 months. The included factors are the market factor and sorted long-short portfolios from the probability entropy and the Kolmogorov-Smirnov measures of ambiguity. Then prices for these factors are estimated using the Fama-MacBeth procedure.

```python
amb_factors = spreading_portfolio_returns
three_factors = factors_monthly['Mkt-RF'].to_frame().merge(amb_factors, on='Date')
```

```python
%%time
data_fmb = pd.DataFrame(index=data_full.index,columns=['return','Mkt-RF','probability_entropy','kolmogorov_smirnov'])
data_fmb['return'] = data_full[('Returns','return')]
for iWin in range(0,len(three_factors.index)-60):
    index = three_factors.index[iWin+60]
    for iInd in industries_monthly.columns:
        y = industries_monthly.loc[three_factors.index,iInd].iloc[iWin:iWin+60].values
        x = three_factors.iloc[iWin:iWin+60,:]
        try:
            betas = np.dot(np.linalg.inv(np.dot(x.T,x)),np.dot(x.T,y))
            data_fmb.loc[(iInd,index),['Mkt-RF','probability_entropy','kolmogorov_smirnov']] = betas
        except:
            pass
    print(index)
```

```python
data_fmb = data_fmb[data_fmb['return'].notna()]
```

```python
fmb = lm.FamaMacBeth(data_fmb['return'],data_fmb[['Mkt-RF','probability_entropy','kolmogorov_smirnov']].shift(1)).fit()#cov_type='kernel')
```

```python
fmb
```

The probability entropy factor has a negative price of similar magnitude as the market factor (-56 bps monthly, or -6.52% annual), however not significant. The price of the Kolmogorov-Smirnov factor is negative and insignificant.

### Conclusion from analysis part
Overall, both tested ambiguity measures repeatedly show a negative relation to subsequent returns. However, the statistical magnitude of this evidence is not compelling.
